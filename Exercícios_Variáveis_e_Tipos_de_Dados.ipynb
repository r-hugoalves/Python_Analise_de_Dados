## **4.Booleanos**

Utilize a tabela da verdade para responder: qual o valor da variável x?

```python
a = False
b = True

x = not a & b
```

O valor é: `true`

Isso porque:

* a `&` b = `false`

Já que para a condição `&` ser verdadeira, ambos devem ser verdadeiros

* `not` (`false`) = `true`
